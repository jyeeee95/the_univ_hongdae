## 이상치 제거

직선거리 카테고리화

```python
dist_arange = []

for index, row in gangnam_temp.iterrows():
    if row["distance"] <= 0.5:
        dist_arange.append("500m")
    elif row["distance"] <= 1:
        dist_arange.append("1000m")
    elif row["distance"] <= 1.5:
        dist_arange.append("1500m")
    elif row["distance"] <= 2:
        dist_arange.append("2000m")
    else:
        dist_arange.append("na")
    
gangnam_temp["dist_arange"] = dist_arange
gangnam_temp.head(20)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


def find_histogram(clt):
    """
    create a histogram with k clusters
    :param: clt
    :return:hist
    """
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)

    hist = hist.astype("float")
    hist /= hist.sum()

    return hist

def plot_colors2(hist, centroids):
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0

    for (percent, color) in zip(hist, centroids):
        # plot the relative percentage of each cluster
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX

    # return the bar chart
    return bar


```

```python
img = cv2.imread("image/gangnam/contour/sampling/gangnam_201703.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = img.reshape((img.shape[0] * img.shape[1],3)) #represent as row*column,channel number
clt = KMeans(n_clusters=6) #cluster number
clt.fit(img)

hist = find_histogram(clt)
print(hist)

bar = plot_colors2(hist, clt.cluster_centers_)

plt.axis("off")
plt.imshow(bar)
plt.show()
```
